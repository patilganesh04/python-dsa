```python
def top_view_tree(root):
    if root is None:
        return []
    hd={}
    queue=[(root,0)]
    while queue:
        node,hdistance=queue.pop(0)
        if hdistance not in hd:
            hd[hdistance]=node.data
        if node.left:
            queue.append((node.left,hdistance-1))
        if node.right:
            queue.append((node.right,hdistance+1))
    min_hd=min(hd.keys())
    max_hd=max(hd.keys())
    result=[]
    for i in range(min_hd,max_hd+1):
        result.append(hd[i])
    return result
```
