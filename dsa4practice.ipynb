```python
class node:
    def __init__(self,value):
        self.value=value
        self.next=None
class singlelinkedlist:
    def __init__(self,head=None):
        self.head=head
    def append(self,value):
        new_node=node(value)
        if self.head is None:
            self.head=new_node
        else:

            curr=self.head
            while curr.next is not None:
                curr=curr.next
            curr.next=new_node
        return self.head

sl=singlelinkedlist()
sl.append(10)
sl.append(20)
sl.append(20)
sl.append(20)
sl.append(20)


curr=sl.head
while curr is not None:
    print(curr.value,end="->")
    curr=curr.next
print("None")
 

 

```

```python
def insert_at_first(head,value):
    new_node=node(value)
    new_node.next=head
    return new_node
sl.head = insert_at_first(sl.head, 0)
sl.head = insert_at_first(sl.head, 10)

# print updated list
curr = sl.head
while curr is not None:
    print(curr.value, end="->")
    curr = curr.next
print("None")

 
```

```python
def insert_middle(self,value,position):
    new_node=node(value)
    if position==0:
        new_node.next=self.head
        self.head=new_node
        return self.head
    curr=self.head
    prev=None
    count=0
    while curr.next is not None and count<position:
        count+=1
        prev=curr
        curr=curr.next
    prev.next=new_node
    new_node.next=curr
    return self.head
sl = singlelinkedlist()
sl.append(10)
sl.append(20)
sl.append(30)
sl.append(40)


sl.insert_middle(5, 0)   # insert 5 at beginning
sl.insert_middle(25, 3)  # insert 25 at index 3

# print updated list
curr = sl.head
while curr is not None:
    print(curr.value, end="->")
    curr = curr.next
print("None")
        

```
