```python
#doubly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    def insert_at_beginning(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
        return self.head
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node
        new_node.prev = curr
        return self.head
    def insert_middle(self, data, position):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        if position == 0:
            return self.insert_at_beginning(data)
        curr = self.head
        prev = None
        count = 0
        while curr and count < position:
            prev = curr
            curr = curr.next
            count += 1
        print(count)
        new_node.next = curr
        new_node.prev = prev
        if curr:
            curr.prev = new_node
        if prev:
            prev.next = new_node
        return self.head
# create doubly linked list
sl = DoublyLinkedList()
sl.insert_at_end(10)
sl.insert_at_end(20)
sl.insert_at_end(30)
sl.insert_at_end(40)
sl.insert_at_beginning(0)
sl.insert_middle(15, 2)
sl.insert_middle(35, 4)
def print_list(head):
    curr = head
    while curr:
        print(curr.data, end=" <-> " if curr.next else "\n")
        curr = curr.next
print_list(sl.head)
```
