```python
def armstrong(num):
    n=len(num)
    total=0

    for i in num:
        id=id**n
        total+=id
    return total
number=armstrong([int(x) for x in input("Enter a number: ")])
if number==int(''.join(input("Enter a number: "))):
    print("The number is an Armstrong number")
print(number)
```

```python
def armstrong_number(n):    
     
    num=n
    total=0
    nid=len(str(n))

    while num>0:
        id=num%10
        total=total+(id**nid)
        num=num//10
    return total
result=armstrong_number(153)
if result==153:
    print("The number is an Armstrong number")
else:
    print("The number is not an Armstrong number")
print(result)
```
