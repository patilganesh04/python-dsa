```python
#quick sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
# Example usage
sample_array = [3,6,8,10,1,2,1]
sorted_array = quick_sort(sample_array)
print(sorted_array)
```

```python
def quick1(arr, low, high):
    if low < high:
        pivot = partition(arr, low, high)
        quick1(arr, low, pivot - 1)
        quick1(arr, pivot + 1, high)

def partition(arr, low, high):
    pivot = arr[low]
    i = low
    j = high

    while i < j:
        while i <= high - 1 and arr[i] <= pivot:
            i += 1
        while j >= low + 1 and arr[j] > pivot:
            j -= 1
        if i < j:
            arr[i], arr[j] = arr[j], arr[i]

    # swap pivot with correct position
    arr[low], arr[j] = arr[j], arr[low]
    return j

# Example usage
sample_array1 = [3, 6, 8, 10, 1, 2, 1]
quick1(sample_array1, 0, len(sample_array1) - 1)
print(sample_array1)
```
