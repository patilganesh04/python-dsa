```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.left = None
        self.right = None

```

```python
def display(root):
    if root is not None:
        print(root.data, end=' ')
        display(root.left)     
        display(root.right)
display(root)
```

```python
#sum of all nodes in Bt
def sum_of_nodes(root):
    if root is None:
        return 0
    return root.data + sum_of_nodes(root.left) + sum_of_nodes(root.right)
print("\nSum of all nodes:", sum_of_nodes(root))
```

```python
def max_tree(root):
    if root is None:
        return float('-inf')
    else:
        left_max = max_tree(root.left)
        right_max = max_tree(root.right)
        return max(root.data, left_max, right_max)
print("Max value in tree:", max_tree(root))
```

```python
def find_height(root):
    if root is None:
        return -1
    else:
        left_height = find_height(root.left)
        right_height = find_height(root.right)
        return 1 + max(left_height, right_height)
print("Height of tree:", find_height(root))
```

```python
def size_of_tree(root):
    if root is None:
        return 0
    else:
        return 1 + size_of_tree(root.left) + size_of_tree(root.right)
print("Size of tree:", size_of_tree(root))
```

```python
#tree traversals

def inorder_traversal(root):
    if root is not None:
        inorder_traversal(root.left)
        print(root.data, end=' ')
        inorder_traversal(root.right)
        
print("Inorder Traversal:", end=' ')
inorder_traversal(root)

def preorder_traversal(root):
    if root is not None:
        print(root.data, end=' ')
        preorder_traversal(root.left)
        preorder_traversal(root.right)

print("\nPreorder Traversal:", end=' ')
preorder_traversal(root)

def postorder_traversal(root):
    if root is not None:
        postorder_traversal(root.left)
        postorder_traversal(root.right)
        print(root.data, end=' ')

print("\nPostorder Traversal:", end=' ')
postorder_traversal(root)
```

```python
 #bfs
class Node:
    def __init__ (self, data):
        self.data = data
        self.left = None
        self.right = None
from collections import deque

def bfs_traversal(root):
    if root is None:
        return
    queue = deque([root])
    while queue:
        current = queue.popleft()
        print(current.data, end=' ')
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

 
 
root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left=Node(4)
root.left.right=Node(5)
print("\nBFS Traversal of new tree:", end=' ')
bfs_traversal(root)
```

```python
#print n th level of tree
from collections import deque
def print_nth_level(root,n):
    if root is None:
        return
    level=0
    queue = deque([root])

    while queue:
        level_size = len(queue)
        if level == n:
            for i in range(level_size):
                current = queue.popleft()
                print(current.data, end=' ')
        else:
            for i in range(level_size):
                current = queue.popleft()
                if current.left:
                    queue.append(current.left)
                if current.right:
                    queue.append(current.right)
        level += 1
print("\nNodes at level 2:", end=' ')
print_nth_level(root,2)

```

```python
#iterative preorder traversal

class Node:
    def __init__ (self, value):
        self.value = value
        self.left = None
        self.right = None

def preorder(root):
    if root is None:
        return
    stack = []
    res=[]
    stack.append(root)
    while stack:
        node = stack.pop()
        res.append(node.value)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return res
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
print("\nIterative Preorder Traversal:", preorder(root))
```

```python
#inorder iterative
def inorder(root):
    stack = []
    res = []
    current = root
    while True:
        if current is not None:
            stack.append(current)
            current = current.left
        elif stack:
            current = stack.pop()
            res.append(current.value)
            current = current.right
        else:
            break
    return res
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
print("\nIterative Preorder Traversal:", preorder(root))
print("\nIterative Inorder Traversal:", inorder(root))
```

```python
#postorder iterative

def postorder(root):
    res=[]
    stack1=[]
    stack2=[]
    if root is None:
        return res    
    stack1.append(root)
    while stack1:
        node=stack1.pop()
        stack2.append(node)
        if node.left:
            stack1.append(node.left)
        if node.right:
            stack1.append(node.right)
    while stack2:
        node=stack2.pop()
        res.append(node.value)
    return res
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
print("\nIterative Postorder Traversal:", postorder(root))
```

```python
#boundary of data
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.data = key

def print_left_boundary(root, res):
    if root:
        if root.left:
            res.append(root.data)
            print_left_boundary(root.left, res)
        elif root.right:
            res.append(root.data)
            print_left_boundary(root.right, res)
#boundary of data --- IGNORE ---
def print_leaves(root, res):
    if root:
        print_leaves(root.left, res)
        if root.left is None and root.right is None:
            res.append(root.data)
        print_leaves(root.right, res)
def print_right_boundary(root, res):
    if root:
        if root.right:
            print_right_boundary(root.right, res)
            res.append(root.data)
        elif root.left:
            print_right_boundary(root.left, res)
            res.append(root.data)

def print_boundary(root):
    if root is None:
        return []
    res = [root.data]
    print_left_boundary(root.left, res)
    print_leaves(root.left, res)
    print_leaves(root.right, res)
    print_right_boundary(root.right, res)
    return res

root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(4)
root.left.right = Node(12)
root.left.right.left = Node(10)
root.left.right.right = Node(14)
root.right.right = Node(25)
print("\nBoundary Traversal of tree:", print_boundary(root))
```

```python
 #level order traversal
def level_order_traversal(root):
    res=[]
    if root is None :
        return res
    queue = deque([root])
    while queue:
        level_size = len(queue)
         
        for i in range(level_size):
            current = queue.popleft()
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
            if i==level_size-1:
                res.append(current.data)
    return res
```
