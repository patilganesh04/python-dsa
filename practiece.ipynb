```python
def reverse(arr):
    left,right=0,len(arr)-1

    while left<right:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1
    return arr
arr=[1,2,3,4,5]
print(reverse(arr))
```

```python
def sorted_array(a):
    n=len(a)

    for i in range(n-1):
        for j in range(n-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a
arr=[5,4,3,2,1]


print(sorted_array(arr))
```

```python
def inser_array(arr):
    n=len(arr)
    for i in range(n):
        min_index=i
        for j in range(n):
            if arr[j]<arr[min_index]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
            
        return arr
arr=[2,1,3,4]
print(inser_array(arr))
```
